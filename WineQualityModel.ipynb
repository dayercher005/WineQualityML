# Import the libraries:

```python
import pandas as pd
import numpy as np
```

# Importing of dataset using Pandas:

```python
wine_data = pd.read_csv("winequality-red.csv")
wine_data.info()
wine_data
```

# Preparation of Data

```python
from sklearn.model_selection import train_test_split

y = wine_data.quality
wine_features = ['pH', 'citric acid', 'alcohol', 'residual sugar', 'sulphates']
X = wine_data[wine_features]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# Constructing Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=1)
# Fit Model
model.fit(X_train, y_train)
# Predicting wine qualities
predictions = model.predict(X_test)

print("Making predictions for the following 5 wine quality: ")
print(X.head())
print("The predictions are: ")
print(model.predict(X.head()))
```

# Validating Model:

```python
from sklearn.metrics import mean_absolute_error

# Calculates mean absolute error on validation data of wine quality.
model_mae = mean_absolute_error(y_test, predictions)

print("Validation MAE for Random Forest Model: {}" .format(model_mae))
```
